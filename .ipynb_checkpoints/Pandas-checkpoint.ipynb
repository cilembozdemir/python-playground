```python
df_obj3=pd.DataFrame(np.arange(36).reshape(6,6))
df_obj3
```

```python
df_obj4=pd.DataFrame(np.arange(15).reshape(5,3))
df_obj4
```

```python
#concenatenating data
df5=pd.concat([df_obj3,df_obj4],axis=1)
```

```python
df5.dropna(axis=0)
```

```python
#drop 
df5.drop([2,5])
```

```python
#drop column write axis=1
df5.drop([0,2],axis=1)
```

```python
series_obj = Series(np.arange(6))
series_obj.name= "added_variable"
series_obj
```

```python
variable_added= DataFrame.join(df5, series_obj)
```

```python
variable_added
```

```python
double_added=variable_added.append(variable_added,ignore_index=True)
double_added
```

```python
#sorting data
data_sorted=double_added.sort_values(by=(5),ascending=False)
data_sorted
```

```python
#read csv files
cars=pd.read_csv('mtcars.csv')
```

```python
cars.head()
```

```python
#naming columns
cars.columns=['carname','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
```

```python
cars.head()
```

```python
#grouping by a column
cars_groups=cars.groupby(cars['cyl'])
cars_groups.mean()
```

```python
answer=cars.groupby(cars['am']).mean()
```

```python
answer
```

```python
#creating subset from maın table
cars_subset=cars[['am','hp']]
cars_subset.head()
```
